Nous avons donc un problème d'équations aux dérivées partielles (EDP) à résoudre sur un domaine d'études. Ces équations  possédent des conditions sur le bord de ce domaine d'études appelées "conditions aux limites".

Trouver Le potentiel électrique V(x) pour 0 < x < L 
<p style="text-align: center;">
$\Delta V(x)=0  \\
avec V(0) = 0 V, V(L) = 1V$
</p>
Si les EDP possèdent un opérateur en temps, des conditions temporelles initiales doivent également être définies. 

L'inconnue de l'EDP est une fonction à définir sur  le domaine d'études avec ces différentes conditions. 
A priori il n'existe généralement pas de formes analytiques de cette fonction. Une méthode numérique peut  être utilisée.  L'adjectif "Numérique" sous entend le recours à l'outil informatique pour trouver la solution de cette EDP. Dans cet optique, le domaine d'étude sur lequel on souhaite résoudre va être maillé (découper avec des éléments géométriques de forme simple (segments, arc de cercle, triangles, quadrangle, tetraedre, hexaedres, prismes). La méthode des éléments finis va nous permettre de définir un système matriciel à partir en utilisant ce maillage. Les inconnues de ce système linéaire seront les valeurs de la fonctions sur certaines entités géométriques du maillage (points, aêtes, faces).  

La solution trouvée sera toujours continue mais reconstruite à partir de ces valeurs discretes interpolées en tout point du domaine d'étude. 
Le  succès de la MEF repose sur trois ingrédients fondamentaux:

+ Les propriétés interpolantes des éléments finis: Ils permettent d'approcher des fonctions sur un domaine en le maillant et en définissant sur chaque maille des fonctions d'interpolation (appellée fonctions de forme). Ces fonctions sont le plus souvent des combinaisons de formes polynomiales et permettent d'interpoler des valeurs connues sur des parties géométriques de la maille (noeuds, arêtes, facettes) en n'importe quel point appartenant à la maille. 
+ Sa programmation est facile, elle  permet en particulier une écriture informatique structurée pouvant utiliser des paragdigmes informatiques poussés comme l'approche objet.\\ 
+ Elle s'appuie sur un cadre théorique solide fournit par l'**Analyse Fonctionnelle**.
 


#### Analyse fonctionnelle

**L'analyse fonctionnelle** est une discipline des mathématiques qui consiste à construire des **espaces vectoriels normés V** complet  sur le corps des nombres réels ou des nombres complexes dans lesquels on va rechercher la fonction qui vérifie notre EDP.
On parle d'espaces fonctionnels. Ces espaces vectoriels sont peuvent être de dimension finie ou infinie.
Ils sont munis d'une **norme**. Cette norme est issues d'un produit scalaire. De tels espaces vectoriels sont dénommés **Espaces de Hilbert**.
Ce produit scalaire  permet de décomposer la fonction recherchée sur une base de cet espace vectoriel.
A ce produit scalaire on associe une norme. Cette norme est généralement homogène à une énergie associée à la physique considérée. 
Lorsque qu'une fonction converge vers une autre fonction au sens de la norme on parle de convergence forte.
Lorsqu'une fonction converge au sens du produit scalaire (composante par composante), on parle de convergence faible.

La résolution de notre EDP, par la MEF nécessite sa réécriture sous forme dite faible, variationnelle ou intégrale. Cette forme peut être vue comme une approche énergétique de notre problème.
Il permet de faire apparaître le produit scalaire associé à notre espace vectoriel et de retrouver la notion de convergence faible.
Notre problème aux EDP peut s'écrire :
\begin{eqnarray}
\left\{ \begin{array}{l}
Chercher \ u \in V  \ tel \ que  \\ \\
a(u,v)=f(v), \ \forall v  \in W
\end{array}\right.
\label{pb_mef}
\end{eqnarray}

W représente l'espace des fonctions "test"- fonctions sur un ouvert de $ R^n $ dans $ R $ indéfiniment dérivables à support compact. 
Et **a(,)** désigne une forme bilinéaire définie sur VxW et **f** une forme linéaire sur W. \\
Un problème  d'EDP est dit bien posé si il admet une unique solution. Si le problème est bien posé la convergence au sens faible est équivallente à la convergence au sens fort. l'Analyse fonctionnelle fournit des outils (théorèmes, lemmes, ...) pour prouver l'existence et l'unicité d'une solution.

####  Discretisation
Le domaine d'étude va être découpé en éléments géométriques de formes simples (discrétisation du domaine d'études). En fonction de la dimension d'espace considérée, on aura des segments en 1D, triangles ou quadrangles en 2D, tétraèdres, hexaèdres, pyramides, prismes en 3D.
On parle de maillage ou pavage.
Une solution discrête va être recherchée sur un espace vectoriel approché $V_h$. $V_h$ est générallement muni de la même norme et du même produit scalaire que V. $ V_h$ est un espace vectoriel de dimension finie.

\begin{eqnarray}
\left\{ \begin{array}{l}
Chercher \ u_h \in V_h  \ tel \ que  \\ \\
a(u_h,v_h)=f(v_h), \ \forall v_h  \in W_h
\end{array}\right.
\label{pb_mef_discret}
\end{eqnarray}

#### Erreur
La norme associée à V va permettre de quantifier l'erreur d'approximation que l'on commet   sur un élement pour passer de la solution exacte à la solution approchée.
Si le problème est bien posé l'erreur entre la solution calculée et la solution exacte doit tendre vers zero, lorsque la taille de l'élément tend vers zero au sens de la norme de de V.
On parle alors **d'estimation d'erreur a priori**. Elle repose sur un maillage de bonne "qualité" que l'on raffine uniformément si le domaine d'étude est important le temps de calcul important pour obtenir une solution avec suffisamment de précision peut être important.

On considère de nouveau le problème  de la chaleur en statique

Trouver T(x) solution du  problème aux limites:


$\frac{k\delta^2 T}{\delta x^2} = 0, \ 0<x<L \\  $ <br />
$T(0)=T_{cold}; T(L)=T_{hot} $

+ T est la température
+ L la longueur entre les deux parois
+ k La conductivité thermique du matériau entre les parois

L'objectif est de résoudre ce problème numériquement avec la méthode des éléments finis. 
La résolution de ce problème par la méthode des éléments finis nécessite son écriture sous forme variationnelle ou forme faible. On montre que résoudre ce problème  équivaut à résoudre le problème ci-dessous :

$\int_{\Omega}{\frac{\delta}{\delta x}(\frac{k\delta T(x)}{\delta x})\Phi(x) \ d\Omega} = 0, \ 0<x<L \\  $ <br />
$T(0)=T_{cold}; T(L)=T_{hot} $



On cherche une fonction T de carré sommable ($\int_{\Omega}{T(x)^2d\Omega}$ l’intégrale existe, fonction à énergie finie) et la dérivée de T est également de carré sommable.

La fonction $\Phi $ est appelée fonction test, elle vérifie les mêmes propriétés que la solution  T aux limites du domaine d’étude.

En intégrant par partie le problème devient
$\int_{\Omega}(\frac{k\delta T(x)}{\delta x}){\frac{\delta \Phi(x)}{\delta x} \ d\Omega}-[\Phi(x)\frac{\delta \Phi}{\delta x}]^L_0 = 0, \ 0<x<L \\  $ <br />

L'intégration sur l'ensemble du domaine s'écrit $\int_{\Omega}(\frac{k\delta T(x)}{\delta x}){\frac{\delta \Phi(x)}{\delta x} \ d\Omega} = \sum_{k=1}^{N}{\int_{e_k}{(\frac{k\delta T(x)}{\delta x}){\frac{\delta \Phi(x)}{\delta x} \ dK}}}$

On recherche une solution numérique approchée Th(x). En éléments finis, cette solution approchée est construite à partir de 2 données :

1. un maillage Mh du domaine de calcul W,
2. un choix d’interpolation Pk sur ce maillage

Pour notre domaine de calcul unidimensionnel $\Omega = [0, L]$, le maillage correspond à un découpage de $\Omega$ en N segments $e_k=[x_i,x_{i+1}]$. Ces segments sont appelés les éléments finis. Les extrémités de chacun de ces segments sont appelés les noeuds du maillage. C’est en ces noeuds que vont être évaluées les températures. On  recherche la température T_i en chaque noeud de ce maillage. 
La matrice A est construite en évaluant les termes intégrales sur chaque élément fini. Le second membre est obtenu à partir du terme source et des conditions aux limites.

Sur chaque segment (élément fini), on choisit une interpolation polynômiale. Cette interpolation permet de calculer la température T(x) en tout point de coordonnée x appartenant à chaque élément à partir des valeurs de la température aux nœuds de l’élément. Le type de l’interpolation dépend du problème à traiter. Pour notre problème elle doit en particulier respecter les points suivants :
1. les intégrales dans la formulation faible  ou variationnelle  doivent exister, sans que le problème dégénère,
2. on cherche d’autre part une solution approchée globalement continue.

La première condition impose que sur chaque élément le polynôme d’interpolation x) soit au moins de degré 1 pour pouvoir calculer les dérivées premières de la solution qui interviennent dans la formulation faible. Sur chaque élément $e_k = [x_k,x_{k+1}]$, la solution approchée est donc un polynôme P(x) de degré l ≥ 1.
Pour un polynôme d’ordre 1 sur [xk,xk+1] (éléments finis P1), on utilise 2 points d’interpolation : les 2 extrémités du segment {S1 = xk, S2 = xk+1} .
